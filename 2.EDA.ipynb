```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cleaned_train.csv")
df.head()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Plot histogram using seaborn for column 2 (index 1)
column_2_name = df.columns[1]
print(column_2_name)
sns.histplot(df[column_2_name], kde=False, color='skyblue', bins=30, ax=axes[0])
axes[0].set_title(f'Histogram of {column_2_name}')
axes[0].set_xlabel(f'{column_2_name} Values')
axes[0].set_ylabel('Frequency')

# Plot histogram using seaborn for column 3 (index 2)
column_3_name = df.columns[2]
sns.histplot(df[column_3_name], kde=False, color='skyblue', bins=30, ax=axes[1])
axes[1].set_title(f'Histogram of {column_3_name}')
axes[1].set_xlabel(f'{column_3_name} Values')
axes[1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
